# Operations

```python
def append_ones(matrix):
    ones_column = np.ones((matrix.shape[0], 1))
    new_matrix = np.concatenate((ones_column, matrix), axis=1)
    return new_matrix
```

```python
def get_inverse(matrix):
    # try:
      inverse = np.linalg.inv(matrix)
      return inverse
    # except np.linalg.LinAlgError:
    #     return None
```

```python
def get_transpose(matrix):
    return np.transpose(matrix)
```

```python
def matrix_multiply(matrix1, matrix2):
    # try:
      result = np.matmul(matrix1, matrix2)
      return result
    # except ValueError:
    #     return None

# Example usage
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
result_matrix = matrix_multiply(matrix1, matrix2)
```

**Prepare the datafram**

```python
df = pd.DataFrame()
for x in range(5):
    y = 6*x + 3
    new_data = {'x': x, 'y': y}
    df = df.append(new_data, ignore_index=True)
```

```python
df
```

```python
column_values = df['y'].tolist()
y = np.array(column_values).reshape(-1, 1)
column_values = df['x'].tolist()
x = np.array(column_values).reshape(-1, 1)
```

```python
x.shape
```

```python
y.shape
```

```python
def MSSW(x,y):
  # x = append_ones(x)
  x_t = get_transpose(x)
  xx_t=matrix_multiply(x_t, x)
  step = get_inverse(xx_t)
  pesodo = matrix_multiply(step,x_t)
  w = matrix_multiply(pesodo, y)
  return w
```

```python
MSSW(x,y)
```

```python
def acc(Y_hat,Y_test):
  c = 0
  for i in range(len(Y_test)):
    if Y_hat[i] > 0 and Y_test[i] == "Iris-setosa":
      c+=1
    elif Y_hat[i] < 0 and Y_test[i] != "Iris-setosa":
      c+=1
  print(c/len(Y_test) *100 , "%" )
```
